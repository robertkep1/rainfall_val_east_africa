**TAMSAT**

```python
tamsat = xr.open_dataset('/content/gdrive/MyDrive/rainfall-forecast/merge_model/ConvLSTM/predictors/predictors_regridded/TAMSAT_regridded.nc')
tamsat = tamsat.loc[dict(time=slice("2016-01-01", "2020-12-31"))]
tamsat = tamsat.fillna(0)
tamsat = tamsat.rfe_filled.values
```

```python
tamsat_min = np.nanmin(tamsat, axis=(1, 2), keepdims=True)
tamsat_max = np.nanmax(tamsat, axis=(1, 2), keepdims=True)
tamsat_min_overall = np.nanmin(tamsat_min) 
tamsat_max_overall = np.nanmax(tamsat_max)
for i in np.arange(len(tamsat_min)):
    tamsat_min[i] = tamsat_min_overall
    tamsat_max[i] = tamsat_max_overall
tamsat = (tamsat - tamsat_min)/(tamsat_max - tamsat_min)
tamsat = np.expand_dims(tamsat, axis = 3)
```

```python
feature_tensor_5d_tamsat = np.empty(shape = (1827-5, 6, 285, 245, 1))
for i in np.arange((1827 - 5)):
  for k in np.arange(6):
        window = tamsat[i:i+6]
        feature_tensor_5d_tamsat[i] = window
np.save('/content/gdrive/MyDrive/rainfall-forecast/merge_model/ConvLSTM/predictors/predictors_regridded/5d_feature_tensors/tamsat_feature_tensor', feature_tensor_5d_tamsat)
```
