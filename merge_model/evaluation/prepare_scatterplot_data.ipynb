```python
import numpy as np
import pandas as pd

from google.colab import drive
drive.mount('/content/gdrive')
```

```python
model_1_obs_daily = pd.read_csv('/content/gdrive/MyDrive/rainfall-forecast/merge_model/evaluation/final_station_data_model1.csv')
model_1_obs_daily = model_1_obs_daily.set_index(['Unnamed: 0'])
model_1_obs_daily = model_1_obs_daily.drop(index = (pd.date_range('2016-01-01','2020-01-03').strftime('%Y-%m-%d')))
model_1_obs_daily_melt = pd.melt(model_1_obs_daily)
model_1_obs_daily_melt.to_csv('/content/gdrive/MyDrive/rainfall-forecast/verification_validation/scatterplot_data_station-model-1.csv')

model_1_sat_daily = pd.read_csv('/content/gdrive/MyDrive/rainfall-forecast/merge_model/evaluation/final_model1.csv')
model_1_sat_daily = model_1_sat_daily.set_index(['Unnamed: 0'])
model_1_sat_daily_melt = pd.melt(model_1_sat_daily)
model_1_sat_daily_melt.to_csv('/content/gdrive/MyDrive/rainfall-forecast/verification_validation/scatterplot_data_merge-model-2.csv')

model_2_obs_daily = pd.read_csv('/content/gdrive/MyDrive/rainfall-forecast/merge_model/evaluation/final_station_data_model2.csv')
model_2_obs_daily = model_2_obs_daily.set_index(['Unnamed: 0'])
model_2_obs_daily = model_2_obs_daily.drop(index = (pd.date_range('2016-01-01','2020-01-03').strftime('%Y-%m-%d')))
model_2_obs_daily_melt = pd.melt(model_2_obs_daily)
model_2_obs_daily_melt.to_csv('/content/gdrive/MyDrive/rainfall-forecast/verification_validation/scatterplot_data_station-model-2.csv')

model_2_sat_daily = pd.read_csv('/content/gdrive/MyDrive/rainfall-forecast/merge_model/evaluation/final_model2.csv')
model_2_sat_daily = model_2_sat_daily.set_index(['Unnamed: 0'])
model_2_sat_daily_melt = pd.melt(model_2_sat_daily)
model_2_sat_daily_melt.to_csv('/content/gdrive/MyDrive/rainfall-forecast/verification_validation/scatterplot_data_merge-model-2.csv')
```

```python
df_scatter_data = pd.DataFrame()
df_scatter_data['model_1_obs_daily'] = model_1_obs_daily_melt['value']
df_scatter_data['model_1_sat_daily'] = model_1_sat_daily_melt['value']
df_scatter_data['model_2_obs_daily'] = model_2_obs_daily_melt['value']
df_scatter_data['model_2_sat_daily'] = model_2_sat_daily_melt['value']
df_scatter_data.to_csv('/content/gdrive/MyDrive/rainfall-forecast/verification_validation/scatterplot_data_merge-model.csv')
```
