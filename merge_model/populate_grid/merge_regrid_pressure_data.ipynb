```python
dataset_1 = '/content/gdrive/MyDrive/rainfall-forecast/merge_model/populate_grid_v2/ERA5/pressure_levels_v2/specific_cloud_liquid_water_content.nc'
dataset1 = nc.Dataset(dataset_1)
```

```python
lats = dataset1.variables['latitude'][:]
lons = dataset1.variables['longitude'][:]
lat_bnds, lon_bnds = [-12.183, 3.622], [28.417, 41.194]
#lat_bnds, lon_bnds = [-12, 3], [28, 42]

lat_inds = np.where((lats > lat_bnds[0]) & (lats < lat_bnds[1]))
lat_inds = np.asarray(lat_inds)

lon_inds = np.where((lons > lon_bnds[0]) & (lons < lon_bnds[1]))
lon_inds = np.asarray(lon_inds)

lat_list = np.ma.getdata(dataset1.variables['latitude'][lat_inds[0,0]:lat_inds[0,-1]])
lon_list = np.ma.getdata(dataset1.variables['longitude'][lon_inds[0,0]:lon_inds[0,-1]])

lat_interp = np.arange(start = lat_list[0], stop = lat_list[-1], step = -0.05)
lon_interp = np.arange(start = lon_list[0], stop = lon_list[-1], step = 0.05)
```

```python
file_names = os.listdir('/content/gdrive/MyDrive/rainfall-forecast/merge_model/populate_grid/ERA5/pressure_levels')
file_names.remove('regridded')
```

```python
for names in file_names:
  print(names)
  keys = list(nc.Dataset('/content/gdrive/MyDrive/rainfall-forecast/merge_model/populate_grid/ERA5/pressure_levels/' + names).variables.keys())[4]
  ds = xr.open_dataset('/content/gdrive/MyDrive/rainfall-forecast/merge_model/populate_grid/ERA5/pressure_levels/' + names)
  ds_int = ds.interp(latitude = lat_interp, longitude = lon_interp, method = 'nearest')
  print('one')
  ds_int.to_netcdf('/content/gdrive/MyDrive/rainfall-forecast/merge_model/populate_grid/ERA5/pressure_levels/regridded/' + names)
```
