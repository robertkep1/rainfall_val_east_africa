```python
# CHIRPS CONVERT TO MONTHLY
obs_daily = pd.read_csv('/content/gdrive/MyDrive/rainfall-forecast/verification_validation/final_station_data_daily_v2_chirps.csv')
obs_daily['Unnamed: 0'] = pd.to_datetime(obs_daily['Unnamed: 0'])
obs_daily = obs_daily.set_index('Unnamed: 0')

obs_monthly = obs_daily.groupby(pd.Grouper(freq='M')).sum()
obs_monthly_notnan = obs_daily.groupby(pd.Grouper(freq='M')).count()
obs_monthly_nan_bool = obs_monthly_notnan == 0
obs_monthly[obs_monthly_nan_bool] = np.nan

obs_monthly.to_csv('/content/gdrive/MyDrive/rainfall-forecast/verification_validation/final_station_data_monthly_v2_chirps.csv')

chirps_daily = pd.read_csv('/content/gdrive/MyDrive/rainfall-forecast/verification_validation/final_chirps_data_daily_v2_corrincl.csv')
chirps_daily['Unnamed: 0.1'] = pd.to_datetime(chirps_daily['Unnamed: 0.1'])
chirps_daily = chirps_daily.set_index('Unnamed: 0.1')

chirps_monthly = chirps_daily.groupby(pd.Grouper(freq='M')).sum()
chirps_monthly_notnan = chirps_daily.groupby(pd.Grouper(freq='M')).count()
chirps_monthly_nan_bool = chirps_monthly_notnan == 0
chirps_monthly[chirps_monthly_nan_bool] = np.nan
chirps_monthly.to_csv('/content/gdrive/MyDrive/rainfall-forecast/verification_validation/final_chirps_data_monthly_v2_corrincl.csv')
```
