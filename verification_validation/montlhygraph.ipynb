```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import plotly.graph_objects as go
import plotly.express as px
from shapely.geometry import Point, Polygon
import os
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
data = pd.read_csv('/content/gdrive/MyDrive/rainfall-forecast/verification_validation/monthlygraphdata.csv')
print(data.columns)
```

```python
fig, ax = plt.subplots(figsize=(15,10))

plt.plot(data.Months,data['Ground Truth'], color = 'green')
plt.fill_between(data.Months, data['GT down'], data['GT up'], color='green', alpha=0.1)

plt.plot(data.Months,data['CHIRPS'],color='gold')
plt.fill_between(data.Months, data['CH down'], data['CH up'], color='gold', alpha=0.1)

plt.plot(data.Months,data['CMORPH'], color = 'tab:blue')
plt.fill_between(data.Months, data['CM down'], data['CM up'], color='tab:blue', alpha=0.1)

plt.plot(data.Months,data['IMERG'], color = 'tab:red')
plt.fill_between(data.Months, data['IM down'], data['IM up'], color='tab:red', alpha=0.1)

plt.plot(data.Months,data['TAMSAT'], color = 'darkgrey')
plt.fill_between(data.Months, data['TA down'], data['TA up'], color='darkgrey', alpha=0.1)

ax.set_xlabel('Months', fontsize=12)
ax.set_ylabel('Precipitation (mm)', fontsize=12)
plt.title('Monthly mean rainfall variability', fontsize = 19)
lg = ax.legend(('Ground truth','CHIRPS','CMORPH','IMERG','TAMSAT'),prop={"size":14},loc=0)

plt.savefig('/content/gdrive/MyDrive/rainfall-forecast/verification_validation/metrics/PNG/monthly/throughtime.png')
```
