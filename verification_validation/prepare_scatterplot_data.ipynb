```python
import numpy as np
import pandas as pd

from google.colab import drive
drive.mount('/content/gdrive')
```

```python
chirps_obs_daily = pd.read_csv('/content/gdrive/MyDrive/rainfall-forecast/verification_validation/final_station_data_daily_v2_chirps.csv')
chirps_obs_daily = chirps_obs_daily.set_index(['Unnamed: 0'])
chirps_obs_daily_melt = pd.melt(chirps_obs_daily)

chirps_sat_daily = pd.read_csv('/content/gdrive/MyDrive/rainfall-forecast/verification_validation/final_chirps_data_daily_v2_corrincl.csv')
chirps_sat_daily = chirps_sat_daily.set_index(['Unnamed: 0.1'])
chirps_sat_daily = chirps_sat_daily.drop(columns=['Unnamed: 0'])
chirps_sat_daily_melt = pd.melt(chirps_sat_daily)

cmorph_obs_daily = pd.read_csv('/content/gdrive/MyDrive/rainfall-forecast/verification_validation/final_station_data_daily_v2_cmorph.csv')
cmorph_obs_daily = cmorph_obs_daily.set_index(['Unnamed: 0'])
cmorph_obs_daily_melt = pd.melt(cmorph_obs_daily)

cmorph_sat_daily = pd.read_csv('/content/gdrive/MyDrive/rainfall-forecast/verification_validation/final_cmorph_data_daily_v2.csv')
cmorph_sat_daily = cmorph_sat_daily.set_index(['Unnamed: 0'])
cmorph_sat_daily_melt = pd.melt(cmorph_sat_daily)

imerg_obs_daily = pd.read_csv('/content/gdrive/MyDrive/rainfall-forecast/verification_validation/final_station_data_daily_v2_imerg.csv')
imerg_obs_daily = imerg_obs_daily.set_index(['Unnamed: 0'])
imerg_obs_daily_melt = pd.melt(imerg_obs_daily)

imerg_sat_daily = pd.read_csv('/content/gdrive/MyDrive/rainfall-forecast/verification_validation/final_imerg_daily_v2.csv')
imerg_sat_daily = imerg_sat_daily.set_index(['Unnamed: 0'])
imerg_sat_daily_melt = pd.melt(imerg_sat_daily)

tamsat_obs_daily = pd.read_csv('/content/gdrive/MyDrive/rainfall-forecast/verification_validation/final_station_data_daily_v2_tamsat.csv')
tamsat_obs_daily = tamsat_obs_daily.set_index(['Unnamed: 0'])
tamsat_obs_daily_melt = pd.melt(tamsat_obs_daily)

tamsat_sat_daily = pd.read_csv('/content/gdrive/MyDrive/rainfall-forecast/verification_validation/final_tamsat_data_daily_v2.csv')
tamsat_sat_daily = tamsat_sat_daily.set_index(['Unnamed: 0'])
tamsat_sat_daily_melt = pd.melt(tamsat_sat_daily)
```

```python
df_scatter_data = pd.DataFrame()
df_scatter_data['chirps_obs_daily'] = chirps_obs_daily_melt['value']
df_scatter_data['chirps_sat_daily'] = chirps_sat_daily_melt['value']
df_scatter_data['cmorph_obs_daily'] = cmorph_obs_daily_melt['value']
df_scatter_data['cmorph_sat_daily'] = cmorph_sat_daily_melt['value']
df_scatter_data['imerg_obs_daily'] = imerg_obs_daily_melt['value']
df_scatter_data['imerg_sat_daily'] = imerg_sat_daily_melt['value']
df_scatter_data['tamsat_obs_daily'] = tamsat_obs_daily_melt['value']
df_scatter_data['tamsat_sat_daily'] = tamsat_sat_daily_melt['value']
df_scatter_data.to_csv('/content/gdrive/MyDrive/rainfall-forecast/verification_validation/scatterplot_data_daily.csv')
```
